```python
x = df["time"]
y = df["D"]
colors = [1, 2, 3]

plt.plot(x, y, color="black")
plt.axvspan(xmin=0, xmax=3.7, color="#62B5E5", alpha=0.3)
plt.axvspan(xmin=3.7, xmax=8, color="#FF9933", alpha=0.3)
plt.axvspan(xmin=8, xmax=50, color="#66FFB2", alpha=0.3)
plt.axvline(x = 8, ymax=0.925, color = "red", linestyle=":", label = 'axvline - full height')
plt.xlim([0, x.max()])
plt.ylim([0, y.max() + 5])
plt.fill_between(x, y, where=(x>=3) & (x<=27), hatch='//', alpha=0.6, facecolor="none")

plt.xlabel("Time (minutes)")
plt.ylabel("Thrombin (nM)")
plt.title("Thrombin Generation Assay")
```
